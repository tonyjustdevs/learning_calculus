## $lim_{x\to0} \frac{1-\cos{x}}{x\sin{x}}$ 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib.ticker import MaxNLocator
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define data
xs = np.linspace(-0.50, 0.5, 51)
fx_cosx = lambda x: np.cos(x)
fx_1 = lambda x: x * 0 + 1
ys_cosx = fx_cosx(xs)
ys_1 = fx_1(xs)

# Plot the lines
plt.plot(xs, ys_1, 'bo-', linewidth=2, markersize=8, label='$f(x)=1$')
plt.plot(xs, ys_cosx, 'r^-', linewidth=2, markersize=8, label='$f(x)=cos(x)$')

# Zoom into the region near zero
plt.xlim(-0.1, 0.1)  # X-axis range
plt.ylim(0.99, 1.01)  # Y-axis range

# Add grid, title, and legend
plt.grid(color='lightgrey', linestyle='--', linewidth=0.5)
plt.title(r"$\lim_{x \to 0}:$ ${x}$ & ${cos x}$", loc='left')
plt.legend(loc='upper right')

# Optionally, add vertical and horizontal lines to highlight the zoomed area
ax = plt.gca()  # Get the current axis
ax.axvline(x=0, color='grey', linestyle='--', linewidth=0.5)
ax.axhline(y=1, color='grey', linestyle='--', linewidth=0.5)
```

```python
# x vs sinx
xs = np.linspace(0,1,20)
fx_sinx = lambda x: np.sin(x)
fx_x    = lambda x: x
ys_sinx = fx_sinx(xs)
ys_x    = fx_x(xs)
plt.plot(xs, ys_x, 'bo--', linewidth=1, markersize=8)
plt.plot(xs, ys_sinx, 'r^--', linewidth=1, markersize=8)
plt.grid(color = 'lightgrey', linestyle = '--', linewidth = 0.5)
plt.title(r"$\lim_{x \to 0}:$ ${x}$ vs ${sin x}$", loc='left')
```

```python

import numpy as np
import matplotlib.pyplot as plt

# X-AXIS VALUES
x_at_c = 0
abt_c = 0.1
xs_n = 51
xs_min = x_at_c - abt_c
xs_max = x_at_c + abt_c
xs = np.linspace(xs_min,xs_max,xs_n)

# X-EXCLUDE LIMIT VALUE
xs = xs[xs != 0]
xs = xs[xs != 0]

# Y-AXIS VALUES
# def fx(x): return 3/(np.sqrt(3*x+1)+1)
fx = lambda x: (1-np.cos(x))/(x*np.sin(x))
ys = fx(xs)

# DRAW PLOTS
plt.scatter(xs, ys, marker="o")
plt.grid(color = 'lightgrey', linestyle = '--', linewidth = 0.5)
# X-LIMIT & VALUE
plt.vlines(x_at_c,linestyles="dotted", ymin=plt.ylim()[0], ymax=max(ys)) # non-monotonic
plt.plot(x_at_c, 0.5,marker="o",markersize=15, markerfacecolor='none', markeredgecolor='red')

# AXIS INCREMENTS
import matplotlib.ticker as ticker
ax = plt.gca()  # Get the current axis
ax.xaxis.set_major_locator(MaxNLocator(nbins=10)) 
ax.yaxis.set_major_locator(MaxNLocator(nbins=10)) 
# TITLE
plt.title(r"$\lim_{x \to 0} \frac{1 - \cos x}{x \sin x}$ [Exercise 2.3.89 Thomas et al., 2014, p. 77]", loc='left')
# plt.title("", loc = 'left')

# OTHER
# # b+-- , o:b , r^ , bo    plt.xlabel("") 
# plt.ylim(bottom=0)  # chart starts from y=0
# ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.000025)) # minor ticks
# ref: https://matplotlib.org/stable/users/explain/axes/axes_ticks.html
```
