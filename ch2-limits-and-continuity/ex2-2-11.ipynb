##### Ex. 11 
$lim_{x\to-3}(x^2-13)$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(-3.1, -2.9, 50)
```

```python
def fx(x): return (x**2-13)
ys = fx(xs)
```

```python

# at x=-3
y_x_at_c = fx(-3)
y_x_at_c
```

```python
plt.scatter(xs,ys)
plt.grid(True)
# help(plt.hlines)
# import matplotlib.pyplot as plt
plt.vlines(x=-3, ymax=100, ymin=-100, linestyles="dotted")
```
