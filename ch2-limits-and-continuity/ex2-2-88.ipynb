## $lim_{x\to3} \frac{x^2-9}{\sqrt{x^2+7}-4}$ 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import math
from matplotlib.ticker import MaxNLocator
```

```python
xs = np.linspace(-20,20,101)
fx = lambda x: (x**2-9)/((np.sqrt(x**2+7)-4))
```

```python
ys = fx(xs)
plt.scatter(xs,ys,marker="+")
plt.grid()
# plt.yticks(np.arange(ys[0],ys[-1],5))
# Ensure ticks are generated within the correct range
# y_min, y_max = ys.min(), ys.max()
# plt.yticks(np.arange(y_min, y_max + 1, 1))  # Adjust step size based on range

ax = plt.gca()  # Get the current axis
ax.xaxis.set_major_locator(MaxNLocator(nbins=10))  # 10 ticks on the x-axis
ax.yaxis.set_major_locator(MaxNLocator(nbins=20))  # 8 ticks on the y-axis
# plt.vlines(x=3, ymin=ys[0], ymax=ys[-1], linestyles="dashed")
plt.vlines(x=3, ymin=min(ys), ymax=max(ys), linestyles="dashed")


```
