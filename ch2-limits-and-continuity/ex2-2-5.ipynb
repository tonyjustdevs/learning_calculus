#### Exercise 2.5: Explain why the limits do or do not exist:

$\lim_{x\to0}\frac{x}{|x|}$

```python
import pandas as pd
import numpy as np
```

```python
xs = np.linspace(-0.1,0.1,1000)
xs = xs[xs != 0] # remove x=0
```

```python
def f(x): return (x)/(np.abs(x))
ys = f(xs)
```

```python
import matplotlib.pyplot as plt
plt.scatter(xs,ys)
plt.grid(True)
```
